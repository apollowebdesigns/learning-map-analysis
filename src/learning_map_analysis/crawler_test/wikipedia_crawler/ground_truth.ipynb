```python
from bs4 import BeautifulSoup as bs
import requests

res = requests.get("https://en.wikipedia.org/wiki/User:David_Eppstein/Graph_Algorithms")
soup = bs(res.text, "html.parser")
naval_battles = {}
for link in soup.find_all("a"):
    url = link.get("href", "")
    if "/wiki/" in url:
        naval_battles[link.text.strip()] = url

print(naval_battles)
```
